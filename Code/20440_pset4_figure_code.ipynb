```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
import io
df = pd.read_excel(io.BytesIO(uploaded['dataset_processed.xlsx']))
```

```python
display(df)
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
a = 0
s = 0
m = 0

for i in range(0,len(df.columns)):
  if df.iloc[0,i] == "Asymptomatic":
    a = a+1
  elif df.iloc[0,i] == "Mild":
    m = m+1
  else:
    s = s+1
Severity = ["Asymptomatic","Mild","Severe"]
values = [a,m,s]
fig, ax = plt.subplots()

bars = plt.bar(Severity, values,
        width = 0.4)
bars[0].set_color('r')
bars[1].set_color('y')
ax.bar_label(bars)
plt.xlabel("Covid-19 Severity", fontweight='bold')
plt.ylabel("No. of Patients",fontweight='bold')
plt.title("Number of patients under each severity group from the dataset")
plt.show()
```
